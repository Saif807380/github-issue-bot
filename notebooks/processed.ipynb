```python
# There are issues that have emojis in them, so removing emojis from them
def de_emojify(text):
    regrex_pattern = re.compile(pattern = "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags = re.UNICODE)
    return regrex_pattern.sub(r'',text)
```

```python
df["title"] = df["title"].apply(de_emojify)
```

```python
df["body"] = df["body"].apply(de_emojify)
```

```python
def remove_non_english(df):
    indices = []
    for i, row in df.iterrows():
        included = False
        try:
            # check if non english detected in body
            for l in Detector(row["body"]).languages:
                if l.name not in ("English", "un") and l.confidence > 0:
                    indices.append(i)
                    included = True
        # if detector is unsure, it throws an exception
        except (UnknownLanguage, Exception):
            indices.append(i)
            included = True
        # if not detected in body check in title
        if not included:
            try:
                for l in Detector(row["title"]).languages:
                    if l.name not in ("English", "un") and l.confidence > 0:
                        indices.append(i)
                        included = True
            except (UnknownLanguage, Exception):
                indices.append(i)
                included = True
        if i%1000 == 0:
            print(f"Rows done = {i}, Length of indices = {len(indices)}")
    return indices 
```

```python
indices = remove_non_english(df)
```

All the issues contain some amount of markdown symbols which need to be removed. Found this piece of code that does it nicely primarily removing html tags

https://gist.github.com/lorey/eb15a7f3338f959a78cc3661fbc255fe

Irony is this message is written in markdown as well

```python
from bs4 import BeautifulSoup
from markdown import markdown
```

```python
def markdown_to_text(markdown_string):
    """ Converts a markdown string to plaintext """

    # md -> html -> text since BeautifulSoup can extract text cleanly
    html = markdown(markdown_string)

    # remove code snippets
    html = re.sub(r'<pre>(.*?)</pre>', ' ', html)
    html = re.sub(r'<code>(.*?)</code >', ' ', html)

    # extract text
    soup = BeautifulSoup(html, "html.parser")
    text = ''.join(soup.findAll(text=True))

    return text
```

```python
df["issue"] = df["issue"].apply(markdown_to_text)
```

### Lower Case -> Remove links -> Remove punctuations -> Remove numbers -> Remove Stopwords

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
```

```python
stop_words = set(stopwords.words('english'))
```

```python
def clean(text):
    text = text.lower()
    text = re.sub(r"http\S+", " ", text)
    text = text.translate(str.maketrans(string.punctuation, ' '*len(string.punctuation)))
    text = text.translate(str.maketrans(string.digits, ' '*len(string.digits)))
    tokens = word_tokenize(text)
    text = " ".join([token for token in tokens if token not in stop_words])
    return text
```

```python
df["issue"] = df["issue"].apply(clean)
```
